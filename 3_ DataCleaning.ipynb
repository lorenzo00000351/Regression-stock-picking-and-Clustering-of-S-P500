La idea es, primero ordenar por ticker y fiscal Date Ending, luego agrupar por ticker y año imputar por bfill().ffill(). Seguidamente tendremos que decidir en cada caso que fecha nos quedamos. No hacerlo todo a la vez porque cada empresa tiene su casuistica.

** solo lo aplico en el caso de que tenga filas duplicadas por año!

```python
duplicated_groups = duplicates[duplicates["count"] > 1][["ticker", "year"]]
```

```python
# Convertir itertuples a una lista de tuplas y crear la máscara
duplicated_tickers_years = list(duplicated_groups.itertuples(index=False, name=None))
mask = df.set_index(['ticker', 'year']).index.isin(duplicated_tickers_years)

# Aplicar transformación solo a los duplicados # tiene sentido que total Revenue a lo largo de los años tenga una tendencia continua
df.loc[mask, "totalRevenue"]=df.groupby(["ticker", "year"])["totalRevenue"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # tiende a cambiar poco año a año ya que se base en politicas contables fijas
df.loc[mask, "depreciation"]=df.groupby(["ticker", "year"])["depreciation"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # similar a depreciation
df.loc[mask, "depreciationAndAmortization"]=df.groupby(["ticker", "year"])["depreciationAndAmortization"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # es una variable mas volatil y depende de ingresos y gastos que cambian cada año
df.loc[mask, "ebitda"]=df.groupby(["ticker", "year"])["ebitda"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # es una variable mas volatil y depende de ingresos y gastos que cambian cada año
df.loc[mask, "netIncome"]=df.groupby(["ticker", "year"])["netIncome"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # el efectivo fluctua segun inversiones deudas y pagos
df.loc[mask, "cashAndCashEquivalentsAtCarryingValue"]=df.groupby(["ticker", "year"])["cashAndCashEquivalentsAtCarryingValue"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # puede variar en funcion de la demanda y producción
df.loc[mask, "inventory"]=df.groupby(["ticker", "year"])["inventory"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # cambian en función de operaciones de corto plazo
df.loc[mask, "currentNetReceivables"]=df.groupby(["ticker", "year"])["currentNetReceivables"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # cambian en función de operaciones de corto plazo
df.loc[mask, "currentAccountsPayable"]=df.groupby(["ticker", "year"])["currentAccountsPayable"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados
df.loc[mask, "totalShareholderEquity"]=df.groupby(["ticker", "year"])["totalShareholderEquity"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # si la empresa no emite nuevas acciones o recompra, debería mantenerse estable
df.loc[mask, "commonStockSharesOutstanding"]=df.groupby(["ticker", "year"])["commonStockSharesOutstanding"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados
df.loc[mask, "shortLongTermDebtTotal"]=df.groupby(["ticker", "year"])["shortLongTermDebtTotal"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # son dependientes de decisiones de inversión y financiamento de la empresa
df.loc[mask, "capitalExpenditures"]=df.groupby(["ticker", "year"])["capitalExpenditures"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # son dependientes de decisiones de inversión y financiamento de la empresa
df.loc[mask, "cashflowFromInvestment"]=df.groupby(["ticker", "year"])["cashflowFromInvestment"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # son dependientes de decisiones de inversión y financiamento de la empresa
df.loc[mask, "dividendPayout"]=df.groupby(["ticker", "year"])["dividendPayout"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados  # son dependientes de decisiones de inversión y financiamento de la empresa
df.loc[mask, "cashflowFromFinancing"]=df.groupby(["ticker", "year"])["cashflowFromFinancing"].transform(
    lambda x: x.bfill().ffill()
    )

# Aplicar transformación solo a los duplicados # es una variable mas volatil y depende de ingresos y gastos que cambian cada año
df.loc[mask, "operatingCashflow"]=df.groupby(["ticker", "year"])["operatingCashflow"].transform(
    lambda x: x.bfill().ffill()
    )
```

```python
# Función para imputar con bfill + ffill por empresa -> para datos que siguen tendencias estables
def imputar_bfill_ffill(df, columnas):
    df[columnas] = df.groupby("ticker")[columnas].transform(lambda x: x.bfill().ffill())

# Función para interpolación lineal por empresa -> para datos volátiles pero con tendencia clara
def imputar_interpolacion(df, columnas):
    df[columnas] = df.groupby("ticker")[columnas].transform(lambda x: x.interpolate(method="linear"))

# Función para imputar con la media por empresa -> datos con fluctuaciones anuales pero sin cambios bruscos
def imputar_media_empresa(df, columnas):
    for col in columnas:
        df[col] = df.groupby("ticker")[col].transform(lambda x: x.fillna(x.mean()))

# Función para imputar con la mediana por empresa -> valores de alta volatilidad
def imputar_mediana_empresa(df, columnas):
    for col in columnas:
        df[col] = df.groupby("ticker")[col].transform(lambda x: x.fillna(x.median()))

# Aplicar estrategias por empresa en df_dropped
imputar_bfill_ffill(df_dropped, [
    "totalRevenue", "depreciationAndAmortization", "totalLiabilities",
    "totalShareholderEquity", "commonStockSharesOutstanding", "totalAssets",
    "totalCurrentAssets", "totalCurrentLiabilities"
])

imputar_interpolacion(df_dropped, ["ebitda", "netIncome", "operatingCashflow"])

imputar_media_empresa(df_dropped, ["cashAndCashEquivalentsAtCarryingValue", "shortLongTermDebtTotal"])

imputar_mediana_empresa(df_dropped, ["capitalExpenditures", "cashflowFromInvestment", "cashflowFromFinancing"])

```
